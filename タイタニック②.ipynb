```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("titanic/train.csv")
test = pd.read_csv("titanic/test.csv")
df
```

```python
df["Survived"].value_counts()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
pd.pivot_table(df, index = "Survived", columns = "Pclass", values = "Age")
```

```python
df.loc[(df["Pclass"] == 1) & (df["Survived"] == 0) & (df["Age"].isnull()), "Age"] = 44
df.loc[(df["Pclass"] == 1) & (df["Survived"] == 1) & (df["Age"].isnull()), "Age"] = 35
df.loc[(df["Pclass"] == 2) & (df["Survived"] == 0) & (df["Age"].isnull()), "Age"] = 34
df.loc[(df["Pclass"] == 2) & (df["Survived"] == 1) & (df["Age"].isnull()), "Age"] = 26
df.loc[(df["Pclass"] == 3) & (df["Survived"] == 0) & (df["Age"].isnull()), "Age"] = 27
df.loc[(df["Pclass"] == 3) & (df["Survived"] == 1) & (df["Age"].isnull()), "Age"] = 21
df.isnull().sum()
```

```python
df.groupby("Sex").mean()["Survived"]
```

```python
df["Embarked"] = df["Embarked"].fillna("S")
```

```python
ohe1 = pd.get_dummies(df["Sex"], drop_first = True)
ohe2 = pd.get_dummies(df["Embarked"], drop_first = True)
```

```python
df2 = pd.concat([df, ohe1, ohe2], axis = 1)
df2
```

```python
df3 = df2.drop(columns = ["PassengerId", "Name", "Sex", "Ticket", "Cabin", "Embarked"])
df3
```

```python
X = df3.drop(columns = ["Survived"])
y = pd.Series(df3["Survived"])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

pipelines = {"rf":Pipeline([("rf", RandomForestClassifier(max_depth = 5, random_state = 0))]), 
            "gb":Pipeline([("scl", StandardScaler()), ("gb", GradientBoostingClassifier(max_depth = 5, random_state = 0))]), 
            "lr":Pipeline([("scl", StandardScaler()), ("lr", LogisticRegression(random_state = 0))]), 
            "kn":Pipeline([("scl", StandardScaler()), ("kn", KNeighborsClassifier())])}

for name, pipeline in pipelines.items():
    pipeline.fit(X_train, y_train)
    
print("学習完了")
```

```python
from sklearn.metrics import accuracy_score

scores = {}
for pipe_name, pipeline in pipelines.items():
    scores[(pipe_name,'train')] = accuracy_score(y_train, pipeline.predict(X_train))
    scores[(pipe_name,'test')] = accuracy_score(y_test, pipeline.predict(X_test))
pd.Series(scores).unstack()
```
