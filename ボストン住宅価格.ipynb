```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_boston

data = load_boston()
df  = pd.DataFrame(data.data, columns = data.feature_names)
y = pd.Series(data.target, name = "y")
df
```

```python
for n in df.columns:
    plt.scatter(df[n], y)
    plt.xlabel(n)
    plt.ylabel("y")
    plt.show()
```

```python
df3 = df2.drop([364, 365, 367, 368, 369, 370, 371, 372], axis = 0)
df3.corr()
```

```python
df3 = df3.reset_index(drop = True)
df3
```

```python
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.linear_model import Ridge, Lasso, LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score

X = df3.drop(["y"], axis = 1)
y = pd.Series(df3.loc[:, "y"])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

pipelines = {
    "lr":Pipeline([("scl", StandardScaler()), ("lr", LinearRegression())]), 
    "rg":Pipeline([("scl", StandardScaler()), ("lr", Ridge())]), 
    "ls":Pipeline([("scl", StandardScaler()), ("ls", Lasso())]), 
    "rf":RandomForestRegressor(max_depth = 3, random_state = 0), 
    "gb1":GradientBoostingRegressor(max_depth = 3, random_state = 0),    
    "gb2":GradientBoostingRegressor(max_depth = 4, random_state = 0), 

}

scores = {}
for name, model in pipelines.items():
    model.fit(X_train, y_train)
    scores[('train',name)]=r2_score(y_train,model.predict(X_train))
    scores[('test',name)]=r2_score(y_test,model.predict(X_test))
display(pd.Series(scores).unstack())

#GB2ãŒè‰¯ã„çµæžœã ãŒã€æ±ŽåŒ–æ€§èƒ½ãŒä½Žãã†ã€‚RFã®æ–¹ãŒè‰¯ã„ã‹ã‚‚
```
